```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
texts = [
    # --- Sports ---
    "The match was exciting and the team played well",
    "The player scored a hat-trick in the game",
    "The coach praised the team's strong defense",
    "The tournament attracted fans from all over the world",
    "The star striker signed a new contract with the club",
    "The final match went into extra time",
    "The goalkeeper made several crucial saves",
    "The championship title was decided in the last minute",
    "The team trained hard before the big game",
    "The athlete broke the national record in the race",

    # --- Politics ---
    "The government passed a new law today",
    "The parliament discussed the budget proposal",
    "The president addressed the nation during the speech",
    "The election results sparked debates across the country",
    "The policy changes aim to improve public healthcare",
    "The senator proposed a new economic reform",
    "The government announced stricter environmental regulations",
    "The mayor met with community leaders to discuss development",
    "The opposition criticized the new tax plans",
    "The political summit brought together global leaders",

    # --- Technology ---
    "The new smartphone has amazing features",
    "The software update improves battery life",
    "The company released an innovative AI-powered device",
    "The laptop offers faster performance with the new processor",
    "The tech conference showcased futuristic gadgets",
    "The cybersecurity system protects users from attacks",
    "The app introduces a redesigned user interface",
    "The robot assistant can perform everyday tasks",
    "The startup is working on advanced machine learning models",
    "The virtual reality headset provides an immersive experience",

    # --- Entertainment ---
    "The actor won an award for best performance",
    "The movie received great reviews from critics",
    "The singer released a new album yesterday",
    "The TV series became a global sensation",
    "The music festival attracted thousands of fans",
    "The director announced a sequel to the popular film",
    "The comedian delivered a hilarious show",
    "The documentary gained praise for its storytelling",
    "The dancer impressed the judges with her routine",
    "The theater play sold out within hours"
]
```

```python
labels = [
    # --- Sports ---
    "Sports","Sports","Sports","Sports","Sports",
    "Sports","Sports","Sports","Sports","Sports",

    # --- Politics ---
    "Politics","Politics","Politics","Politics","Politics",
    "Politics","Politics","Politics","Politics","Politics",

    # --- Technology ---
    "Technology","Technology","Technology","Technology","Technology",
    "Technology","Technology","Technology","Technology","Technology",

    # --- Entertainment ---
    "Entertainment","Entertainment","Entertainment","Entertainment","Entertainment",
    "Entertainment","Entertainment","Entertainment","Entertainment","Entertainment"
]
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, label, test_size = 0.4, random_state = 42, stratify = labels)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("classification report: ",classification_report(y_test, y_pred))
print("accuracy score: ",round(accuracy_score(y_test, y_pred*100,2),"%")
```
